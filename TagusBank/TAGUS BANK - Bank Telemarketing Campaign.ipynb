**Context**

Perbankan di Portugal memainkan peran penting dalam perekonomian dengan menyediakan berbagai layanan keuangan, termasuk produk investasi seperti *term deposit*. Produk ini memberikan keuntungan bagi bank dalam menjaga stabilitas keuangan serta menawarkan opsi investasi berisiko rendah bagi nasabah. Namun, konversi pelanggan ke produk ini masih menjadi tantangan, terutama dalam strategi pemasaran yang dilakukan melalui telemarketing.

Tagus Bank menggunakan strategi telemarketing untuk menawarkan produk term deposit kepada nasabah. Meskipun metode ini memungkinkan bank untuk menjangkau lebih banyak nasabah, efektivitasnya masih diragukan karena banyak panggilan yang tidak menghasilkan konversi. Berdasarkan data, dari total **39,971 nasabah yang dihubungi**, hanya **4,486 nasabah yang menerima tawaran term deposit**, yang menghasilkan **Conversion Rate** sebesar **11%**. Angka ini menunjukkan bahwa sebagian besar nasabah menolak penawaran, menandakan bahwa kampanye telemarketing yang dilakukan belum sepenuhnya efektif dalam meningkatkan konversi. Faktor-faktor seperti status pekerjaan, usia, riwayat kampanye sebelumnya, dan durasi panggilan diduga memiliki pengaruh yang signifikan terhadap keputusan nasabah untuk menerima atau menolak penawaran tersebut.

Masih banyak nya tingkat penolakan ini mengindikasikan bahwa strategi telemarketing saat ini belum optimal dalam meningkatkan *Conversion Rate*. Untuk meningkatkan konversi di kampanye berikut nya, diperlukan pendekatan berbasis data guna mengidentifikasi pola dan karakteristik nasabah yang lebih mungkin menerima penawaran. Dengan memahami faktor-faktor utama yang memengaruhi keputusan nasabah, tim telemarketing dapat menyusun strategi yang lebih tepat sasaran dan meningkatkan peluang konversi.

Oleh karena itu, **Head of Telemarketing** di Tagus Bank meminta tim Data Scientist untuk melakukan analisis data dan membangun model prediktif guna mengoptimalkan strategi pemasaran dan meningkatkan rasio konversi telemarketing. Model ini diharapkan dapat membantu tim telemarketing dalam menentukan kelompok nasabah dengan probabilitas konversi lebih tinggi, sehingga upaya pemasaran dapat lebih efektif dan terarah.

Target:  
- 0: Nasabah menolak penawaran term deposit
- 1: Nasabah menerima penawaran term deposit

**Problem Statement**

Strategi telemarketing yang tidak terarah dapat menyebabkan upaya pemasaran kurang efektif, menghabiskan waktu tim telemarketing, dan menurunkan peluang konversi. Saat ini, Tagus Bank menghadapi tantangan dalam meningkatkan conversion rate term deposit, karena kampanye telemarketing masih dilakukan secara luas tanpa mempertimbangkan karakteristik atau riwayat interaksi nasabah.

Jika strategi telemarketing tetap dilakukan secara acak tanpa analisis berbasis data, beberapa permasalahan yang dapat terjadi meliputi:
    
- Conversion Rate dari team Telemarketing tidak meningkat karena aktifitas telemarketing tidak difokuskan ke dataset yang memiliki potensial closing yang lebih besar.
    
- Aktifitas telemarketing yang tidak terarah (Spray and Prey) akan mengakibatkan konversi Nasabah tidak dapat diprediksi dengan baik, sehingga sales Forecast di team Telemarketing akan kurang tepat dan tidak dapat dikontrol.

Dengan adanya analisis yang memahami pola nasabah yang lebih responsif terhadap penawaran serta metode komunikasi yang paling efektif, team telemarketing bisa memprioritaskan aktifitas cold-calling menggunakan metode komunikasi yang lebih efektif dan ke kelompok Nasabah yang memiliki conversion rate tinggi.

Berdasarkan Dataset yang digunakan, sebagian besar Nasabah yang dihubungi team Telemarketing adalah nasabah baru.

**Metric Evaluation**

![Alt text](https://raw.githubusercontent.com/PurwadhikaDev/AlphaGroup_JC_DS_OL_16_B_FinalProject/main/metrics.png)

Dalam konteks telemarketing term deposit, kesalahan prediksi dapat membawa konsekuensi yang signifikan:

- Type 1 Error (False Positive)

  **Konsekuensi**: Tim telemarketing membuang waktu dan sumber daya untuk menghubungi nasabah yang sebenarnya tidak berminat, yang dapat meningkatkan biaya operasional tanpa meningkatkan conversion rate.

- Type 2 Error (False Negative)

  **Konsekuensi**: Kehilangan calon nasabah yang sebenarnya tertarik terhadap term deposit, sehingga berpotensi menurunkan jumlah konversi

Dikarenakan tujuan utama kita adalah meningkatkan efisiensi strategi telemarketing dengan menargetkan nasabah yang benar-benar memiliki potensi konversi tinggi, maka kita perlu meminimalkan False Positives agar tim telemarketing tidak menghubungi nasabah yang tidak tertarik.

Oleh karena itu, **metrik utama yang akan kita gunakan adalah Precision** agar memastikan bahwa sebagian besar nasabah yang dihubungi memiliki potensi konversi tinggi sehingga tim telemarketing dapat fokus pada nasabah yang lebih mungkin menerima penawaran. Namun, kita juga tetap mempertimbangkan trade-off antara Precision dan Recall. Maka kita menggunakan **F0.5-Score sebagai metrik pendukung** karena F0.5 lebih menekankan Precision dibanding Recall, **memastikan bahwa efisiensi tetap tinggi sambil mempertahankan jumlah konversi yang optimal**.

#### **Unknown Value**

Unknown value secara umum terlihat di beberapa kolom diawal penjelasan dataset, ini perlu di cek lanjut apakah missing value input berbeda atau memang mempunyai konteks.

```python
listItem = []

for col in master.columns:
    if master[col].dtype == 'object':
        unique_vals = master[col].value_counts()
        unique_pct = (unique_vals / len(master[col])) * 100
        
        
        for val, count, percentage in zip(unique_vals.index, unique_vals.values, unique_pct.values):
            if val == 'unknown':
                listItem.append([col, master[col].dtype, val, count, round(percentage, 2)])

masterDesc = pd.DataFrame(columns=['dataFeatures', 'dataType', 'uniqueValue', 'counts', 'percentage'],
                          data=listItem)

masterDesc = masterDesc.sort_values(by='percentage', ascending=False)

masterDesc
```

#### **Nasabah Lama dan Nasabah Baru**

##### `previous = 0`, `pdays = 999` dan `poutcome = 'Nonexistent'`

Sebelumnya, kita telah membahas bahwa ketiga kolom ini saling berhubungan. Nasabah baru berarti nasabah yang belum pernah dihubungi sebelumnya, memiliki `previous = 0`, `pdays = 999`, dan `poutcome = 'nonexistent'`. Sebaliknya, nasabah lama sudah pernah dihubungi sebelumnya (`previous > 0`) akan memiliki `pdays` dengan nilai selain 999, serta `poutcome` bernilai `'failure'` atau `'success'`, tergantung hasil kampanye sebelumnya.

Konsisten nya data nasabah diperlukan untuk dapat memvalidasi apakah nasabah baru atau nasabah lama dalam analisis nanti nya. Serta dikarenakan nilai 999 merupakan nilai mayoritas pada kolom pdays, angka yang besar dan banyak ini dapat memengaruhi distribusi data dan deteksi outlier. Oleh karena itu, kita akan melakukan pengecekan untuk memastikan bahwa `pdays` = 999 harusnya berisi `previous = 0` dengan `poutcome = 'nonexistent'`.

```python
filtered_data = master[master['pdays'] == 999][['previous', 'pdays', 'poutcome']]
count_previous_poutcome = filtered_data.groupby(['previous', 'poutcome']).size().reset_index(name='Count')
count_previous_poutcome.insert(0, 'Pdays', 999)
count_previous_poutcome
```

Terlihat pada tabel di atas bahwa `poutcome = 'nonexistent'` dan `previous = 0` tidak hanya memiliki jumlah yang sama, tetapi juga berasal dari baris data nasabah yang sama. Dengan demikian, `poutcome = 'nonexistent'` merupakan kategori valid yang menunjukkan bahwa nasabah belum pernah dihubungi sebelumnya dalam kampanye sebelumnya.  

Selanjutnya, dapat kita lihat bahwa jumlah baris dengan `pdays = 999` lebih banyak, dan sebagian besar nilainya konsisten dengan `previous = 0` serta `poutcome = 'nonexistent'`, sesuai dengan konteks data. Namun, terdapat beberapa kasus di mana nasabah memiliki `pdays = 999` tetapi juga memiliki riwayat dihubungi sebelumnya (`previous > 0`) serta hasil kampanye sebelumnya (`poutcome ≠ 'nonexistent'`).  

Untuk menangani baris yang tidak konsisten tersebut tanpa mengganggu pola distribusi data dan analisis outliers, nilai `pdays = 999` pada kasus yang bermasalah akan diganti dengan modus (nilai yang paling sering muncul, selain 999) berdasarkan kelompok dengan `previous` yang sama.

```python
pdays_summary = (
    master[master["previous"] <= 7]
    .groupby("previous")["pdays"]
    .agg(lambda x: list(x.unique()))
    .reset_index()
)

pdays_summary["count"] = pdays_summary["previous"].apply(
    lambda prev: list(master[master["previous"] == prev]["pdays"].value_counts().sort_index().values)
)

pdays_summary
```

Diatas, terdapat 35,551 baris yang sudah konsisten. Berikut detail baris yang tidak konsisten.

```python
mask_invalid_pdays = (master["pdays"] == 999) & (master["previous"] > 0)
invalid_rows = master[mask_invalid_pdays][["previous", "pdays", "poutcome"]]

num_invalid_rows = invalid_rows.shape[0]
total_rows = master.shape[0]
percentage_invalid = (num_invalid_rows / total_rows) * 100

print(f"Jumlah baris bermasalah: {num_invalid_rows} dari {total_rows} ({percentage_invalid:.2f}%)")

invalid_rows.sort_values(by="previous")
```

Terdapat 4,110 dari 41,176 baris atau 10% baris tidak konsisten yang terindikasi kesalahan input. Untuk itu kita akan lihat unique values pdays dari tiap baris ini:

```python
valid_pdays = master[master["pdays"] != 999]
mode_pdays_per_previous = valid_pdays.groupby("previous")["pdays"].agg(lambda x: x.mode()[0]).reset_index()

mode_pdays_table = pd.DataFrame({
    "previous": mode_pdays_per_previous["previous"],
    "pdays_mode": mode_pdays_per_previous["pdays"]
})
mode_pdays_table
```

Ditemukan nilai modus (mengecualikan pdays 999) dari tiap previous nya. Angka-angka inilah yang akan menggantikan nilai pdays 999 yang tidak konsisten.

```python
previous_to_mode_pdays = dict(zip(mode_pdays_table["previous"], mode_pdays_table["pdays_mode"]))

mask_replace = (master["pdays"] == 999) & (master["previous"] > 0)
master.loc[mask_replace, "pdays"] = master.loc[mask_replace, "previous"].map(previous_to_mode_pdays)
```

```python
remaining_invalid = master[(master["pdays"] == 999) & (master["previous"] > 0) & (master["poutcome"] != 'nonexistent')]
print(f"Sisa baris bermasalah setelah perbaikan: {remaining_invalid.shape[0]}")
```

```python
pdays_summary = (
    master[master["previous"] <= 7]
    .groupby("previous")[["pdays","poutcome"]]
    .agg(lambda x: list(x.unique()))
    .reset_index()
)

pdays_summary["count"] = pdays_summary["previous"].apply(
    lambda prev: list(master[master["previous"] == prev]["pdays"].value_counts().sort_index().values)
)

pdays_summary
```

Kita telah menangani nilai `pdays = 999` tanpa membuang data, hanya mengoreksi kesalahan. Dengan ini sudah dapat dibedakan mana nasabah baru dan nasabah lama. Nilai `999` pada pdays adalah kode khusus yang menandakan "belum pernah dihubungi", bukan angka sebenarnya, karena itu, kita menggantinya dengan **modus** berdasarkan distribusi `pdays` dalam `previous`, sehingga karakteristik kelompok tetap terjaga.  

Walau sudah konsisten antara nasabah baru dan nasabah lama, namun keberadaan `999` bisa menyebabkan bias dalam distribusi angka karena bukan nilai numerik dari distribusi sebenarnya, melainkan kode kategori khusus. Disisi lain kita tetap membutuhkan angka ini untuk menjaga informasi kolom lain karena jumlah baris nya banyak agar tetap bisa mengeksplorasi distribusi serta outlier. #### **Data Distribution**

Selanjutnya, kita akan memeriksa distribusi data pada setiap kolom numerik untuk memahami karakteristik data secara menyeluruh. Langkah ini penting untuk mengidentifikasi potensi ketidakseimbangan dan skewness yang dapat memengaruhi kinerja beberapa algoritma Machine Learning tertentu, terutama yang sensitif terhadap pola distribusi data.

```python
numerical_columns = master.select_dtypes(include=['float64', 'int64']).columns
selected_columns = numerical_columns

for column in selected_columns:
    stat, pval = shapiro(master[column].dropna())
    
    if pval < 0.05:
        print(f'Kolom {column}: Tidak berdistribusi normal')
    else:
        print(f'Kolom {column}: Berdistribusi normal')
```

Semua kolom numerikal tidak berdistribusi normal maka dari itu kita menggunakan median bukan mean untuk melihat pusat data.

```python
numerical_columns = master.select_dtypes(include=['float64', 'int64']).columns

n_cols = 2
n_rows = int(np.ceil(len(numerical_columns) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 4 * n_rows), sharex=False, sharey=False)
axes = axes.flatten()

for i, column in enumerate(numerical_columns):
    median_value = master[column].median()
    
    sns.histplot(
        master[column].dropna(), 
        kde=True, 
        bins=30, 
        ax=axes[i],
        color='blue'
    )
    
    axes[i].axvline(median_value, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')
    
    axes[i].set_title(f'Distribution of {column}', fontsize=12)
    axes[i].set_xlabel(column, fontsize=10)
    axes[i].set_ylabel('Frequency', fontsize=10)
    axes[i].tick_params(axis='both', which='major', labelsize=9)
    axes[i].legend(fontsize=9)

for j in range(len(numerical_columns), len(axes)):
    fig.delaxes(axes[j])

# Refine layout
plt.suptitle("Histograms of Numerical Columns", fontsize=16, y=1.02)
plt.tight_layout(pad=2)
plt.show()
```

```python
print('------------------------------------------------------------------------')
# General Statistic Info untuk kolom age
profit_descriptive = master['age'].describe()
# Menentukan Area Q1 & Q3
Q1 = master['age'].quantile(0.25)
Q3 = master['age'].quantile(0.75)
IQR = Q3 - Q1
# Menentukan batas bawah dan atas untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Menampilkan hasil
print(f'General Statisctic Info: \n{profit_descriptive}')
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

#### **Data Correlation**

Setelah memahami distribusi data dan mengidentifikasi outlier, langkah selanjutnya adalah menganalisis hubungan antar kolom, baik satu sama lain maupun dengan target variabel. 
Analisis hubungan kolom ini bertujuan untuk memastikan bahwa hanya menggunakan kolom relevan sebagai Fitur Variable dan telah bersih untuk pemodelan Machine Learning Classification, memiliki hubungan yang erat dengan Target Variabel, serta independen satu sama lain agar model dapat bekerja secara optimal. Selain itu, dengan memahami hubungan antar fitur ini, kita dapat menangani nilai unik seperti 'unknown' dengan mempertimbangkan domain knowledge serta hubungan teknisnya dalam dataset.

Fitur Variable merupakan kolom-kolom yang digunakan untuk memprediksi Target Variabel, yaitu kolom `y`, dengan kategori sebagai berikut:  
- **0 (No)**: Nasabah menolak tawaran deposito  
- **1 (Yes)**: Nasabah menerima tawaran deposito

Untuk itu kita akan melakukan Label Encoding data didalam kolom `y` agar menjadi numerikal untuk di proses hubungan nya dengan kolom lain. Hanya untuk keperluan analisis maka proses ini menggunakan dataset yang di duplikasi agar data sebelumnya tetap aman, karena encoding akan dilakukan pada Preprocessing untuk data yang sudah bersih.

```python
# Inisialisasi LabelEncoder
le = LabelEncoder()
   
# Ubah kolom y menjadi int
master['y'] = le.fit_transform(master['y'])
```

**Feature Variable Numerical or Multikolinearitas**

Mukltikolinearitas adalah situasi yang menunjukkan adanya korelasi kuat antara dua variabel bebas atau lebih dalam sebuah model regressi. Sebelum mengecek multikolinearitas (hubungan antar variable), kita menganalisis korelasi antar fitur menggunakan Spearman correlation, karena data tidak berdistribusi normal. Tabel di bawah ini memberikan insight tentang hubungan antar variabel, menunjukkan fitur mana yang berkorelasi kuat atau lemah.

```python
# Pilih kolom numerik
numerical_columns = tempp.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Hitung korelasi Spearman
correlation_spearman = tempp[numerical_columns].corr(method='spearman')

# Plot heatmap
correlation_spearman
```

Namun, korelasi tidak cukup untuk mendeteksi multikolinearitas dalam model. Oleh karena itu, kita menggunakan Variance Inflation Factor (VIF) untuk mengevaluasi seberapa signifikan sebuah variabel terpengaruh oleh variabel lainnya dalam model. Jika VIF terlalu tinggi, maka fitur tersebut memiliki multikolinearitas yang tinggi dan mungkin perlu dihapus atau ditransformasikan agar model tetap stabil.

```python
# Pilih hanya fitur numerik (kecuali 'y')
numerical_columns = tempp.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Buat DataFrame untuk VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = numerical_columns
vif_data["VIF"] = [variance_inflation_factor(tempp[numerical_columns].values, i) for i in range(len(numerical_columns))]

# Tampilkan hasil
vif_data.sort_values(by="VIF", ascending=True)
```

Kesimpulan:

- Fitur dengan VIF rendah

    Fitur ini memiliki VIF rendah hingga moderat, tidak memiliki masalah multikolinearitas dan tetap bisa digunakan dalam model karena memiliki hubungan independent dengan variabel lain.

    - `campaign`
    - `duration`
    - `previous`
    - `age`

- Kolom dengan VIF tinggi

    Berdasarkan nilai Variance Inflation Factor (VIF) yang tinggi, kolom-kolom ini harus ditangani lebih lanjut.
    - Kolom `nr.employed`, `cons.price.idx`, `emp.var.rate` dan`euribor3m` memiliki informasi yang tumpang tindih tentang kondisi ekonomi makro, sehingga dapat mempertimbangkan memilih kolom yang paling relevan saja dan perlu dilakukan eksperimen dengan menghapus fitur ini dan melihat dampaknya terhadap model machine learning
    - Kolom `pdays` dari Correlation Spearman dan nilai VIF memiliki multikolinearitas tinggi dan sangat mirip dengan `previous`, ini  benar secara logika telemarketing dimana jika pdays = 999 dan previous = 0 maka nasabah tidak pernah dihubungi pada kampanye sebelumnya.

**Featue Variables Category**

```python
# Pilih hanya kolom kategorikal
categorical_cols = tempp.select_dtypes(include=['object']).columns

# Simpan hasil uji chi-square
chi_square_results = []

# Iterasi setiap pasangan kolom kategorikal
for i in range(len(categorical_cols)):
    for j in range(i + 1, len(categorical_cols)):  # Hindari pengulangan yang sama
        cat1, cat2 = categorical_cols[i], categorical_cols[j]
        
        # Buat tabel kontingensi
        contingency_table = pd.crosstab(temp[cat1], temp[cat2])
        
        # Lakukan uji Chi-Square
        chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
        
        # Simpan hasil
        chi_square_results.append((cat1, cat2, chi2_stat, p_value))

# Konversi hasil ke DataFrame
chi_square_df = pd.DataFrame(chi_square_results, columns=['Category 1', 'Category 2', 'Chi-Square Statistic', 'p-value'])

# Urutkan hasil berdasarkan Chi-Square Statistic (descending) dan p-value (ascending)
chi_square_df_sorted = chi_square_df.sort_values(by=['Chi-Square Statistic', 'p-value'], ascending=[False, True])

# Tampilkan hasil yang telah diurutkan
chi_square_df_sorted
```

Konsep Tabel Korelasi:
- Chi-Square Statistic:
    
    Semakin tinggi nilainya, semakin besar kemungkinan ada hubungan antara kedua variabel.
- p-value:
    - p < 0.05 → Ada hubungan signifikan antara kedua variabel.
    - p ≥ 0.05 → Tidak ada hubungan yang signifikan.

Dari hasil uji Chi-Square antar variabel kategorikal, ditemukan beberapa pasangan yang memiliki hubungan yang sangat signifikan (p-value = 0.0). Ini menunjukkan bahwa ada keterkaitan kuat di antara beberapa fitur.

Fitur yang memiliki hubungan kuat satu sama lain dapat menyebabkan multikolinearitas dalam model berbasis regresi. Jika menggunakan model berbasis pohon keputusan nantinya, fitur yang berkaitan masih bisa digunakan karena model ini dapat menangani korelasi antar fitur.

- **pdays 0-5**: 1 Weeks

- **pdays 11-15**: 3 Weeks  
- **pdays 16-20**: 4 Weeks  
- **pdays 21-25**: 5 Weeks  
- **pdays 26-30**: 6 Weeks
- **pdays >30 dan 999**: Not Contacted  

```python
# Fungsi untuk mengkategorikan pdays
def categorize_pdays(pdays):
    if pdays <= 5:
        return "1 Week"
    elif 6 >= pdays <= 10:
        return "2 Weeks"
    elif 11 >= pdays <= 15:
        return "3 Weeks"
    elif 16 >= pdays <= 20:
        return "4 Weeks"
    elif 21 >= pdays <= 25:
        return "5 Weeks"
    elif 26 >= pdays <= 30:
        return "6 Weeks"
    else:
        return "Not Contacted"

# Terapkan fungsi ke kolom pdays
master['pdays'] = master['pdays'].apply(categorize_pdays)
```

```python
# Pengecekan Kategori
master['pdays'] = master['pdays'].astype(str).astype('object')
master['pdays'].value_counts()
```

Di atas, kita telah mengubah nilai numerik di kolom `pdays` menjadi kategori per minggu. Tujuannya adalah agar pada tahap preprocessing nanti, data ini dapat diproses menggunakan **Ordinal Encoding** sehingga sekarang nilai 999 sudah sebagai kategori yang disebut "Not Contacted".

```python
# Total keseluruhan
total_no = age_conversion_table.loc['Total Keseluruhan', 'Menolak Tawaran']
total_yes = age_conversion_table.loc['Total Keseluruhan', 'Menerima Tawaran']
total_all = total_no + total_yes  # Total semua data

# Persentase untuk NO dan YES
no_percentage = (total_no / total_all) * 100
yes_percentage = (total_yes / total_all) * 100

# Ambil persentase per kelompok usia dari total NO dan YES
no_percentages = age_conversion_table.iloc[:-1]['Menolak Tawaran (%)']
yes_percentages = age_conversion_table.iloc[:-1]['Menerima Tawaran (%)']

# Gabungkan data untuk plotting
all_labels = list(no_percentages.index) + list(yes_percentages.index)
all_counts = no_percentages.tolist() + yes_percentages.tolist()

# Warna yang sama untuk kategori usia dalam YES & NO
colors = {
    '15-24 (Early Working Age)': '#ff9999',  # Merah Muda
    '25-54 (Prime Working Age)': '#66b3ff',  # Biru
    '55-64 (Mature Working Age)': '#99ff99',  # Hijau
    '65+ (Elderly)': '#ffcc99'  # Orange
}

# Generate warna yang sama untuk YES dan NO
pie_colors = [colors[age] for age in no_percentages.index] + [colors[age] for age in yes_percentages.index]

# Buat explode hanya untuk bagian "YES" supaya lebih menonjol
explode = [0] * len(no_percentages) + [0.15] * len(yes_percentages)  # YES meledak sedikit

# Buat Pie Chart
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts = ax.pie(
    all_counts, labels=None, autopct=None,  # Tidak menampilkan autopct langsung
    colors=pie_colors, explode=explode, startangle=140,
    wedgeprops={"linewidth": 0, "edgecolor": "white"} 
)

# Tambahkan persentase (%) di luar pie chart
for i, wedge in enumerate(wedges):
    ang = (wedge.theta2 - wedge.theta1) / 2 + wedge.theta1 # Posisi tengah wedge
    x = 1.2 * np.cos(np.radians(ang))  # Koordinat X di luar pie
    y = 1.2 * np.sin(np.radians(ang))  # Koordinat Y di luar pie
    ax.text(x, y, f"{all_counts[i]:.1f}%", 
            ha="center", va="center", fontsize=10, fontweight="bold")

# Tambahkan label besar di dalam pie untuk menandai YES dan NO
ax.text(0.1, -0.4, f"NO\n{no_percentage:.1f}%", fontsize=20, fontweight='bold', color='black', ha='center', va='center')
ax.text(-0.35, 0.7, f"YES\n{yes_percentage:.1f}%", fontsize=20, fontweight='bold', color='black', ha='center', va='center')

# Tambahkan legend untuk kategori usia di bawah chart
legend_labels = list(colors.keys())
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[age], markersize=10) for age in colors.keys()]
plt.legend(handles, legend_labels, title="Age Category", loc="lower center", bbox_to_anchor=(0.5, -0.15), ncol=2)

# Tambahkan judul
plt.title("Pengaruh Umur Nasabah terhadap Penawaran Term Deposito")

# Tampilkan plot
plt.show()
```

**Bagaimana efektivitas strategi kampanye saat ini dalam meningkatkan konversi penawaran term deposit?**

Untuk memahami efektivitas strategi kampanye saat ini dalam meningkatkan konversi penawaran term deposit, kita dapat melihat Conversion Rate (CR), yang mengukur seberapa efektif kampanye telemarketing dalam menghasilkan keputusan positif (terima penawaran).

$$
\text{Conversion Rate (CR)} = \frac{\text{Jumlah Nasabah yang Menerima Penawaran (y == 'yes')}}{\text{Total Nasabah yang DiHubungi}}
$$

##### **Metode Komunikasi terhadap Convertion Rate Penawaran Term Deposit**

```python
contact_conversion_table = cleaned_data.groupby('contact')['y'].value_counts().unstack().fillna(0)

contact_conversion_table.columns = ['Menolak Tawaran', 'Menerima Tawaran']

total_menolak = contact_conversion_table['Menolak Tawaran'].sum()
total_menerima = contact_conversion_table['Menerima Tawaran'].sum()

contact_conversion_table['Conversion Rate (%)'] = (
    (contact_conversion_table['Menerima Tawaran'] / 
     (contact_conversion_table['Menolak Tawaran'] + contact_conversion_table['Menerima Tawaran'])) * 100
).round(2)

contact_conversion_table
```

```python
conversion_counts = contact_conversion_table[['Menolak Tawaran', 'Menerima Tawaran']]

conversion_percentage = contact_conversion_table[['Conversion Rate (%)']]

colors = ['#ff9999', '#66b3ff']

fig, ax = plt.subplots(figsize=(10, 6))

conversion_counts.plot(kind='bar', stacked=True, color=colors, ax=ax)

for i, (index, row) in enumerate(conversion_percentage.iterrows()):
    bottom = 0
    for col, value in zip(conversion_counts.columns, row):
        height = conversion_counts.loc[index, col]
        if height > 0:
            ax.text(i, bottom + height * 1.2, f"{value:.1f}%", ha='center', va='center', 
                    fontsize=10, color='black')
        bottom += height

ax.set_title('Metode Komunikasi terhadap Convertion Rate Penawaran Term Deposit', fontsize=14, fontweight='bold')
ax.set_xlabel('Jenis Kontak')
ax.set_ylabel('Jumlah Nasabah')

ax.set_xticks(np.arange(len(conversion_counts.index)))
ax.set_xticklabels(conversion_counts.index, rotation=0, fontsize=12)

ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Berdasarkan plot diatas, metode kontak melalui *cellular* memiliki tingkat konversi yang lebih tinggi dibandingkan *telephone*. Dari total kontak *cellular*, 14.69% menerima tawaran, sementara pada *telephone*, hanya 5.27% yang menerima. Hal ini menunjukkan bahwa komunikasi melalui ponsel lebih efektif dalam menarik respon positif dibandingkan telepon rumah atau kantor. Meskipun jumlah penolakan tetap lebih tinggi di kedua metode, *cellular* memberikan peluang yang lebih besar untuk konversi.

#### **Preprocessing**

Ada dua hal yang kita lakukan di Preprocessing Pipeline untuk tranformasi kolom yaitu Encoding dan Feature Scalling. Kedua hal ini sudah disinggung sebelumnya saat eksplorasi data.

Dalam mengubah tipe data kolom menjadi numerik, ada 3 pilihan Encoding:
- One Hot Encoding
    - Hanya berlaku di kolom yang Category nya nominal atau level nya sama
    - Efektif jika 5 data saja yang di Encoding agar menghindari Overfitting
    - Jika lebih dari 5 data, tidak efektif karena One Hot Encoding akan menghasilkan kolom baru sesuai degan banyak nya unique data di kolom
    - Dimana pada Data Cardinalities sebelumnya sudah kita bahas kolom yang menggunakan tipe encoding ini, yaitu:
        * `marital`
        * `default`
        * `poutcome`
        * `day_of_week`

- Binary Encoding
    - Sama seperti One Hot Encoding (data yang nominal atau level nya sama) namun untuk data yang lebih dari 5
    - Cara kerjanya juga sama, akan menghasilkan kolom baru tapi satu kolom saja dengan langsung encoding nya (tidak satuan angka)
    - Dimana pada Data Cardinalities sebelumnya sudah kita bahas kolom yang menggunakan tipe encoding ini, yaitu:
        * `job`
        * `month`

    - Khusus kolom ini akan menggunakan Binary Encoding juga dikarenakan memiliki dua kategori dan tidak menambah kolom baru
        * `contact`

- Ordinal Encoding
    - Digunakan untuk data category yang berurutan/bertingkat. Contoh: SD, SMP, SMA
    - Langsung Encoding data tanpa bikin kolom baru
    - Dimana pada Data Cardinalities sebelumnya sudah kita bahas kolom yang menggunakan tipe encoding ini, yaitu:
        * `education`
        * `pdays`


Dalam pemrosesan fitur numerik, Feature Scaling diperlukan agar model Machine Learning dapat memahami skala data dengan lebih baik. Jika tidak dilakukan scaling, fitur dengan nilai besar dapat mendominasi perhitungan model dan menyebabkan bias dalam prediksi.  

Pada dataset ini, distribusi semua fitur numerik tidak normal dan cenderung mengandung outliers, sehingga metode RobustScaler digunakan. Berbeda dengan StandardScaler yang bergantung pada mean dan standard deviation, RobustScaler menggunakan median dan Interquartile Range (IQR), sehingga lebih tahan terhadap nilai ekstrem dan lebih sesuai untuk data yang memiliki distribusi skewed.  

Fitur numerik yang di-scaling menggunakan RobustScaler meliputi:  
- `campaign`
- `previous`
- `age`
- `emp.var.rate`  
- `euribor3m`  
- `nr.employed`  

```python
# Definisikan transformer dengan Encoding dan RobustScaler
transformer = ColumnTransformer([
    # One Hot Encoding
    ('OneHotEncoder', OneHotEncoder(drop='first', sparse_output=False), [
        'marital', 'default', 'poutcome', 'day_of_week'
    ]),

    # Ordinal Encoding
    ('OrdinalEncoder', OrdinalEncoder(categories=[
        # Urutan tingkat pendidikan dari terendah ke tertinggi
        ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 
         'professional.course', 'university.degree'],
        
        # Urutan pdays berdasarkan minggu
        ['Not Contacted', '1 Week', '2 Weeks', '3 Weeks', '4 Weeks', '5 Weeks', '6 Weeks']
    ]), ['education', 'pdays']),
    
    # Binary Encoding
    ('BinaryEncoder', BinaryEncoder(), ['job', 'month', 'contact']),
    
    # Feature Scaling dengan RobustScaler
    ('RobustScaler', RobustScaler(), ['campaign', 'age', 'previous', 'emp.var.rate', 'euribor3m', 'nr.employed'])
], remainder='passthrough')

```

Jika tidak ditangani, model yang sensitif terhadap imbalance data dapat cenderung hanya memprediksi kelas mayoritas, mengurangi akurasi pada kelas minoritas, serta menurunkan efektivitas strategi telemarketing.

Maka dari itu pada benchmarking model menggunakan Cross Validation, kita akan menggunakan `Stratified K-Fold CV` agar distribusi target tetap proporsional dalam setiap fold. Juga dilangkah selanjutnya kita akan bereksperimen dengan teknik Balancing Data agar menemukan teknik penyeimbangan dengan performa terbaik.

-  Model Individu  
    1. Decision Tree (DT) merupakan berbasis pohon keputusan yang sederhana, cepat, dan dapat menangkap hubungan non-linear dalam data.  
    2. K-Nearest Neighbors (KNN) adalah algoritma berbasis tetangga terdekat yang bekerja dengan membandingkan kemiripan antar data.  
    3. Logistic Regression (LogReg) merupakan model linier yang sering digunakan dalam klasifikasi biner dengan interpretasi probabilitas yang jelas.  
    4. Naïve Bayes (NB) adalah model probabilistik yang efektif untuk data dengan asumsi independensi fitur.  
    5. Random Forest (RF) adalah Ensemble berbasis banyak pohon keputusan, meningkatkan stabilitas dan mengurangi overfitting.  
    6. XGBoost (XGB) merupakan model berbasis boosting yang dikenal dengan performanya yang kuat dan efisien terhadap data yang kompleks.  
    7. Gradient Boosting (GB) adalah algoritma boosting yang membangun model secara bertahap dengan memperbaiki kesalahan dari model sebelumnya.  

- Ensemble Learning

  Ensemble Learning adalah metode menggabungkan beberapa model individu untuk mendapatkan prediksi yang lebih stabil dan akurat.  

  8. Hard Voting atau disebut *majority rule* adalah algoritma gabungan yang menggunakan mayoritas suara dari model individu untuk menentukan prediksi akhir. Disini terdiri dari model Decision Tree (DT) , Random Forest (RF) dan Naive Bayes (NB). Hal itu dikarenakan kombinasi model pohon dan probabilistik memberikan keseimbangan antara interpretabilitas dan kecepatan prediksi.  

  9. Soft Voting atau disebut *probabilistic averaging* memiliki cara kerja dengan mengambil rata-rata probabilitas prediksi dari masing-masing model. Pada metode ini terdapat Logistic Regression, Random Forest, XGBoost dan Gradient Boosting. Dikarenaka kombinasi model yang linear, pohon keputusan, dan boosting memberikan cakupan yang lebih luas terhadap pola data yang kompleks.  

  10. Stacking atau *meta-Learning* menggunakan beberapa model sebagai 'base learners', lalu hasilnya diproses oleh 'meta-model 'untuk membuat prediksi akhir. Pada metode ini digunakan Decision Tree (DT) , Random Forest (RF), XGboost dan Gradient Boosting sebagai base learners nya, yang mana Logistic Regression menjadi meta model nya. Hal itu dikarenakan model yang ada di base learners menangkap berbagai aspek data, sementara Logistic Regression sebagai meta-model menyatukan informasi dari semua model dengan bobot yang optimal.

#### **Eksperimen Balancing**

##### **Class Weight**

Target Variable yang tidak seimbang membuat model bisa jadi bias ke kelas mayoritas, karena jika model selalu memprediksi "No", maka model akan tetap dapat akurasi tinggi. Maka dari itu Class weight dipakai untuk menangani ketidakseimbangan kelas tanpa menambah jumlah sampel baru, dengan teknik ini mengubah cara model menghitung loss, jadi kesalahan di kelas minoritas dihukum lebih berat.

Sederhana nya, tanpa class weight model lebih mudah belajar dari kelas mayoritas (karena jumlahnya lebih banyak). Dengan class weight, model "dipaksa" untuk lebih memperhatikan kelas minoritas, sehingga prediksi lebih seimbang

- Menggunakan class_weight='balanced' untuk model yang mendukung, yaitu:
    - Logistic Regression
    - Decision Tree Classifier
    - Random Forest Classifier

- Tidak menggunakan class weight
    - Gradient Boosting

- Menggunakan scale_pos_weight untuk XGBoost
    - scale_pos_weight = (jumlah kelas negatif / jumlah kelas positif)

- Tetap menggunakan teknik Cross Validation sebelumnya agar tetap konsisten

```python
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}

cw_logreg = LogisticRegression(class_weight='balanced', random_state=2024)
cw_dt = DecisionTreeClassifier(max_depth=5, class_weight='balanced', random_state=2024)
cw_rf = RandomForestClassifier(n_estimators=300, max_depth=10, criterion='gini', 
                               min_samples_split=3, class_weight='balanced', random_state=2024)
cw_xgb = XGBClassifier(scale_pos_weight=class_weights[1] / class_weights[0], random_state=2024)
gb = GradientBoostingClassifier(n_estimators=300, random_state=2024)

cw_soft_vote = VotingClassifier([
    ('LogReg', cw_logreg),
    ('RF', cw_rf),
    ('XGB', cw_xgb),
    ('GB', gb)
], voting='soft')

base_model = ([
    ('DT', cw_dt),
    ('RF', cw_rf),
    ('XGB', cw_xgb),
    ('GB', gb)
])
est_model = LogisticRegression(class_weight='balanced', random_state=2024)
cw_stacking = StackingClassifier(estimators=base_model, final_estimator=est_model)

cw_top_5_models = {
    'RF': cw_rf,
    'LogReg' : cw_logreg,
    'Soft Vote': cw_soft_vote,
    'Stacking': cw_stacking,
    'DT' : cw_dt
}

cw_results = []

for model_name, model in cw_top_5_models.items():
    model_pipe = Pipeline(steps=[
        ('Preprocess', transformer),
        ('Model', model)
    ])
    
    f05_scores = cross_val_score(model_pipe, X_train, y_train, cv=cv, scoring=f0_5_scorer)
    precision_scores = cross_val_score(model_pipe, X_train, y_train, cv=cv, scoring='precision')
    recall_scores = cross_val_score(model_pipe, X_train, y_train, cv=cv, scoring='recall')
    
    cw_results.append({
        'Model': model_name,
        'F0.5-Score': f05_scores.tolist(),
        'AVG F0.5-Score': f05_scores.mean(),
        'Std F0.5-Score': f05_scores.std(),
        'Precision': precision_scores.tolist(),
        'AVG Precision': precision_scores.mean(),
        'Std Precision': precision_scores.std(),
        'Recall': recall_scores.tolist(),
        'AVG Recall': recall_scores.mean(),
        'Std Recall': recall_scores.std(),
    })

cw_df = pd.DataFrame(cw_results)
cw_df
```

##### **SMOTE / Oversampling Balancing**

Selanjutnya kita bereksperimen dengan SMOTE (Synthetic Minority Over-sampling Technique). Teknik ini akan melakukan oversampling pada data latih sebelum melatih model. Hanya saja di teknik balancing ini kita menggunakan `ImbPipeline` bukan `Pipeline` biasa untuk mentransformasi data. Hal ini dikarenakan Pipeline biasa dari sklearn akan menyebabkan data leakage karena SMOTE diterapkan ke seluruh dataset (termasuk test fold dalam cross-validation).

```python
smote = SMOTE(random_state=2024)

smote_top_5_models = {
    'RF': rf,
    'LogReg' : logreg,
    'Soft Vote': soft_vote,
    'Stacking': stacking,
    'DT' : dt
}

smote_results = []

for model_name, model in smote_top_5_models.items():
    smote_pipeline = ImbPipeline(steps=[
        ('Preprocess', transformer),
        ('SMOTE', smote),
        ('Model', model)
    ])
    
    f05_scores = cross_val_score(smote_pipeline, X_train, y_train, cv=cv, scoring=f0_5_scorer)
    precision_scores = cross_val_score(smote_pipeline, X_train, y_train, cv=cv, scoring='precision')
    recall_scores = cross_val_score(smote_pipeline, X_train, y_train, cv=cv, scoring='recall')
    
    smote_results.append({
        'Model': model_name,
        'F0.5-Score': f05_scores.tolist(),
        'AVG F0.5-Score': f05_scores.mean(),
        'Std F0.5-Score': f05_scores.std(),
        'Precision': precision_scores.tolist(),
        'AVG Precision': precision_scores.mean(),
        'Std Precision': precision_scores.std(),
        'Recall': recall_scores.tolist(),
        'AVG Recall': recall_scores.mean(),
        'Std Recall': recall_scores.std(),
    })

smote_df = pd.DataFrame(smote_results)
smote_df
```

```python
smote_df.sort_values('AVG Precision', ascending = False)
```

##### **SMOTE TOMEK Balancing**

Jika sebelumnya kita menggunakan balancing method SMOTE, di tahap ini kita bereksperimen dengan SMOTE-Tomek, yaitu kombinasi dari Synthetic Minority Over-sampling Technique (SMOTE) dan Tomek Links. SMOTE melakukan oversampling pada kelas minoritas dengan mensintesis data baru, sedangkan Tomek Links menghapus sampel yang berdekatan dari kelas mayoritas untuk meningkatkan pemisahan kelas. Teknik ini membantu menyeimbangkan data tanpa overfitting akibat duplikasi sampel.  

Sama seperti sebelumnya, kita menggunakan `ImbPipeline` alih-alih `Pipeline` biasa dari `sklearn`, karena jika menggunakan Pipeline standar, SMOTE akan diterapkan ke seluruh dataset (termasuk test fold dalam cross-validation), yang dapat menyebabkan data leakage dan hasil evaluasi yang tidak valid. Dengan `ImbPipeline`, oversampling hanya diterapkan pada data latih, memastikan evaluasi model tetap akurat.

```python
smote_tomek = SMOTETomek(random_state=2024)

smote_tomek_top_5_models = {
    'RF': rf,
    'LogReg' : logreg,
    'Soft Vote': soft_vote,
    'Stacking': stacking,
    'DT' : dt
}

smote_tomek_results = []

for model_name, model in smote_tomek_top_5_models.items():
    smote_tomek_pipeline = ImbPipeline(steps=[
        ('Preprocess', transformer),
        ('SMOTE-Tomek', smote_tomek),
        ('Model', model)
    ])
    
    f05_scores = cross_val_score(smote_tomek_pipeline, X_train, y_train, cv=cv, scoring=f0_5_scorer)
    precision_scores = cross_val_score(smote_tomek_pipeline, X_train, y_train, cv=cv, scoring='precision')
    recall_scores = cross_val_score(smote_tomek_pipeline, X_train, y_train, cv=cv, scoring='recall')
    
    smote_tomek_results.append({
        'Model': model_name,
        'F0.5-Score': f05_scores.tolist(),
        'AVG F0.5-Score': f05_scores.mean(),
        'Std F0.5-Score': f05_scores.std(),
        'Precision': precision_scores.tolist(),
        'AVG Precision': precision_scores.mean(),
        'Std Precision': precision_scores.std(),
        'Recall': recall_scores.tolist(),
        'AVG Recall': recall_scores.mean(),
        'Std Recall': recall_scores.std(),
    })

smote_tomek_df = pd.DataFrame(smote_tomek_results)
smote_tomek_df
```

#### **Overfitting/Ideal/Underfitting**

Penting untuk memastikan bahwa model telah mencapai kondisi "ideal" sebelum melakukan Hyperparameter Tuning. Tujuan utamanya adalah agar tuning tidak hanya menghafal noise atau pola yang tidak relevan, melainkan dapat menghasilkan model yang mampu menggeneralisasi dengan baik.

Oleh karena itu, langkah pertama adalah membandingkan performa prediksi antara Training set dan Test set. Jika terdapat perbedaan signifikan, ini mengindikasikan kemungkinan overfitting. Sebaliknya, jika kedua performa memiliki nilai rendah, model cenderung underfitting dan perlu dituning untuk meningkatkan akurasi. Model yang ideal akan menunjukkan kestabilan performa di kedua set, dengan nilai metrics yang konsisten dan lebih dari 0.5.

Berdasarkan eksperimen sebelumnya, kita telah menetapkan:
- Random Forest dan Logistic Regression sebagai dua model terbaik untuk benchmarking
- Semua kolom fitur awal tetap dipertahankan
- Penggunaan Cross Validation dengan `Stratified K-Fold 5` untuk evaluasi performa yang lebih reliabel.

```python
models = {
    'RF': rf,
    'LogReg': logreg
}

results = []

for name, model in models.items():
    idealPipeline = Pipeline([
        ('Preprocess', transformer),
        ('Model', model)
    ])

    train_f05 = cross_val_score(idealPipeline, X_train, y_train, 
                               cv=StratifiedKFold(n_splits=5), scoring=f0_5_scorer)
    test_f05 = cross_val_score(idealPipeline, X_test, y_test, 
                              cv=StratifiedKFold(n_splits=5), scoring=f0_5_scorer)
    
    train_precision = cross_val_score(idealPipeline, X_train, y_train, 
                                      cv=StratifiedKFold(n_splits=5), scoring='precision')
    test_precision = cross_val_score(idealPipeline, X_test, y_test, 
                                     cv=StratifiedKFold(n_splits=5), scoring='precision')
    
    train_recall = cross_val_score(idealPipeline, X_train, y_train, 
                                   cv=StratifiedKFold(n_splits=5), scoring='recall')
    test_recall = cross_val_score(idealPipeline, X_test, y_test, 
                                  cv=StratifiedKFold(n_splits=5), scoring='recall')

    results.append({
        'Model': name,
        'Train F0.5-score (CV Mean)': train_f05.mean(),
        'Test F0.5-score (CV Mean)': test_f05.mean(),
        'F0.5 Difference': abs(train_f05.mean() - test_f05.mean()),  
        'Train Precision (CV Mean)': train_precision.mean(),
        'Test Precision (CV Mean)': test_precision.mean(),
        'Precision Difference': abs(train_precision.mean() - test_precision.mean()),
        'Train Recall (CV Mean)': train_recall.mean(),
        'Test Recall (CV Mean)': test_recall.mean(),
        'Recall Difference': abs(train_recall.mean() - test_recall.mean())
    })

idealCheck_result_ = pd.DataFrame(results)
```

Berdasarkan hasil eksperimen, perbandingan performa model Random Forest (RF) dan Logistic Regression (LogReg) ini akan dianalisis menggunakan metrik Precision, F0.5-score dan Recall pada data Train dan Test.

Hasil pengujian menunjukkan bahwa Logistic Regression memiliki performa yang lebih baik dibandingkan Random Forest. Hal ini ditunjukkan oleh nilai F0.5-score pada test data yang lebih tinggi (0.4765) dibandingkan dengan train data (0.4551), dengan selisih 0.0214. Selain itu, precision model Logistic Regression juga meningkat pada test data, dari 0.6638 di train menjadi 0.7004 di test, yang mengindikasikan bahwa model mampu menggeneralisasi data *unseen* dengan baik.

#### **Hyperparameter Tuning**

Di tahap ini kita akan meningkatkan performa model Logistic Regression yang sudah dipilih. Untuk panduan tuning model dapat merujuk ke [Tuning Logistic Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

Sebelum melanjutkan, perlu dilakukan perubahan pada tahap preprocessing menggunakan ColumnTransformer. Pada kolom `default`, yang berisi dua kategori unik yaitu 'yes' dan 'no' diketahui bahwa tidak terdapat data 'yes' pada Train set. Akibatnya, jika ColumnTransformer tidak diperbarui, saat diterapkan pada Test set, berpotensi error karena Train set tidak mengenal nilai 'yes' yang ada di Test set.

```python
print(X_train['default'].value_counts())
print(X_test['default'].value_counts())
```

```python
marital_categories = sorted(X_train['marital'].unique())  
default_categories = ['no', 'yes']
poutcome_categories = sorted(X_train['poutcome'].unique())  
day_of_week_categories = sorted(X_train['day_of_week'].unique()) 
```

Maka akan dilakukan perubahan pada ColumnTransformer dibawah ini.

```python
transformer = ColumnTransformer([
    ('OneHotEncoder', OneHotEncoder(drop='first', sparse_output=False, 
                                    categories=[marital_categories, 
                                                default_categories,  
                                                poutcome_categories, 
                                                day_of_week_categories]),  
     ['marital', 'default', 'poutcome', 'day_of_week']),

    ('OrdinalEncoder', OrdinalEncoder(categories=[
        ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 
         'professional.course', 'university.degree'],
        ['Not Contacted', '1 Week', '2 Weeks', '3 Weeks', '4 Weeks', '5 Weeks', '6 Weeks']
    ]), ['education', 'pdays']),

    ('BinaryEncoder', BinaryEncoder(), ['job', 'month', 'contact']),

    ('RobustScaler', RobustScaler(), ['campaign', 'age', 'previous', 'emp.var.rate', 'euribor3m', 'nr.employed'])
], remainder='passthrough')
```

```python
tuningPipeline = Pipeline([
        ('Preprocess', transformer),
        ('Model', logreg)
])
```

```python
param_dist = {
    'Model__penalty': ['l1', 'l2'],
    'Model__C': uniform(0.01, 10),
    'Model__max_iter': [100, 200, 300, 500]
}
```

```python
random_search = RandomizedSearchCV(
    tuningPipeline,
    param_distributions=param_dist,
    n_iter=100,
    cv=cv,
    scoring='precision',
    random_state=2024,
    n_jobs=-1,
    verbose=1
)
```

```python
random_search.fit(X_train, y_train)
```

Menyimpan model dan parameter nya dalam bentuk pipeline ke variable baru untuk digunakan selanjutnya.

```python
best_model = random_search.best_estimator_
```

```python
print("Best Parameters:", random_search.best_params_)
print("Best Precision Score:", random_search.best_score_)
```

Dari hasil RandomizedSearchCV di atas, diperoleh kombinasi hyperparameter terbaik untuk Logistic Regression sebagai berikut:

- C: 0.1749  
- Max Iter: 300  
- Penalty: L2  

Dengan parameter tuning ini, nilai metric Precision terbaik yang diperoleh adalah 0.6745, yang menunjukkan **peningkatan signifikan dibandingkan dengan baseline sebelumnya**. Meskipun model masih dapat untuk terus dituning, hasil ini menunjukkan bahwa model telah menunjukkan performa yang lebih baik, terutama dengan penyesuaian nilai C dan penggunaan penalty L2 untuk regularisasi yang lebih optimal.

Dalam konteks perhitungan Conversion Rate yang akan kita bahas, confusion matrix akan menunjukkan seberapa baik model kita dalam memprediksi keputusan nasabah apakah mereka menerima atau menolak penawaran term deposit.

Berikut adalah confusion matrix yang dihasilkan:

```python
cm = confusion_matrix(y_test, y_pred_tuned)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Nasabah Menolak', 'Nasabah Menerima'],
            yticklabels=['Nasabah Menolak', 'Nasabah Menerima'])
plt.xlabel('PREDICTED')
plt.ylabel('ACTUAL')
plt.title('Matrix')
plt.show()
```

Detail Confusion Matrix:
- True Negatives (TN): 10,425 Nasabah yang tidak menerima penawaran dan diprediksi tidak menerima penawaran
- False Positives (FP): 138 Nasabah yang tidak menerima penawaran, tetapi diprediksi menerima penawaran
- False Negatives (FN): 1,088 Nasabah yang menerima penawaran, tetapi diprediksi tidak menerima penawaran
- True Positives (TP): 287 Nasabah yang menerima penawaran dan diprediksi menerima penawaran

Skor metrics Precision di model sebelumnya itu dihitung dengan confusion matrix di atas, yaitu:
$$
\text{Precision} = \frac{TP}{TP + FP} = \frac{287}{287 + 138} = 0.67
$$

Ini berarti bahwa dari semua nasabah yang diprediksi menerima penawaran (prediksi positif), 67% atau 287 nasabah benar-benar menerima penawaran (sebenarnya positif), sementara 33% atau 138 nasabah adalah kesalahan prediksi (False Positives).

Dengan implementasi model machine learning, kita menghitung Conversion Rate dengan fokus pada nasabah yang benar-benar menerima penawaran dan diprediksi menerima penawaran dengan benar, dibandingkan dengan total nasabah yang benar-benar menerima penawaran (baik yang diprediksi benar maupun salah). Perhitungannya adalah sebagai berikut:

$$
\text{Conversion Rate (CR)} = \frac{\text{TP}}{\text{TP+FN}}
$$

$$
\text{Conversion Rate (CR)} = \frac{\text{287)}}{\text{287+1088}} = 0.21
$$

Dari perhitungan di atas, **Conversion Rate setelah penerapan model machine learning adalah 21%, yang menunjukkan peningkatan sebesar 10%** dibandingkan dengan conversion rate sebelumnya, yang hanya 11%.
